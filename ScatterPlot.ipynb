```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

# Đọc và vẽ từng file CSV
file_pattern = 'TT T*.csv'
file_list = glob.glob(file_pattern)

for file_name in file_list:
    # Đọc dữ liệu từ file CSV
    data = pd.read_csv(file_name)

    # Lấy dữ liệu từ các cột
    x = data['product_id']
    y = data['net_price']
    colors = data['colors']

    # Vẽ scatter plot với màu sắc được mã hóa
    plt.scatter(x, y, c=colors, cmap='viridis', edgecolors='k', alpha=0.7, label=file_name)

# Thêm colorbar để hiển thị giá trị tương ứng với màu sắc
plt.colorbar()

# Hiển thị biểu đồ
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
from matplotlib.cm import get_cmap

# Đọc và vẽ từng file CSV
file_pattern = 'TT T*.csv'
file_list = glob.glob(file_pattern)

# Chọn colormap và số lượng màu sắc
cmap = get_cmap('viridis')
num_colors = len(file_list)

for i, file_name in enumerate(file_list):
    # Tạo một plot mới cho mỗi file
    plt.figure()

    # Đọc dữ liệu từ file CSV
    data = pd.read_csv(file_name)

    # Lấy dữ liệu từ các cột
    x = data['product_id']
    y = data['net_price']

    # Chọn màu từ colormap
    color = cmap(i / num_colors)

    # Vẽ scatter plot với màu sắc được mã hóa
    plt.scatter(x, y, color=color, edgecolors='k', alpha=0.7, label=file_name)

    # Thêm colorbar để hiển thị giá trị tương ứng với màu sắc
    plt.colorbar()

    # Đặt tiêu đề và nhãn trục
    plt.title(f"Scatter Plot - {file_name}")
    plt.xlabel("Product ID")
    plt.ylabel("Net Price  ")

    # Hiển thị biểu đồ
    plt.legend()
    plt.show()
```
