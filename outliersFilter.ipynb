```python
import pandas as pd
import glob
import os

#file_pattern = 'python_code\Sales_public_csv\TT T*.csv'
#file_list = glob.glob(file_pattern)
sales_split4_folder_path = r"C:\Users\lemin\OneDrive - VLG\Hoc tap\Datathon\code\python_code\Sales_split4_csv"
sales_split4_files = [f for f in os.listdir(sales_split4_folder_path) if f.endswith('.csv')]

# Lấy thư mục của file mã nguồn Python
#current_directory = r"C:\Users\lemin\OneDrive - VLG\Hoc tap\Datathon\code\python_code\Sales_split4_csv"


def remove_outliers_iqr(dataframe, column_name, threshold=1.5):
    """
    Loại bỏ outlier từ DataFrame sử dụng phương pháp IQR.

    Parameters:
        - dataframe: DataFrame chứa dữ liệu
        - column_name: Tên cột chứa dữ liệu cần loại bỏ outlier
        - threshold: Ngưỡng IQR để xác định outlier (mặc định là 1.5)

    Returns:
        DataFrame mới đã loại bỏ outlier
    """
    q1 = dataframe[column_name].quantile(0.25)
    q3 = dataframe[column_name].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - threshold * iqr
    upper_bound = q3 + threshold * iqr

    filtered_data = dataframe[(dataframe[column_name] >= lower_bound) & (dataframe[column_name] <= upper_bound)]

    return filtered_data

for file_name in sales_split4_files:
    read_path = os.path.join(sales_split4_folder_path, file_name)
    output_file_path = os.path.join(sales_split4_folder_path, file_name)

    # Đọc dữ liệu từ file CSV
    df = pd.read_csv(read_path)
    print("Loading and reading data from",file_name)

    # Loại bỏ outlier từ cột 'net_price' và thông báo kqua so sánh
    filtered_df = remove_outliers_iqr(df, 'net_price')
    print(file_name)
    print("Before:", len(df['net_price']))
    print("After:" , len(filtered_df['net_price']))

    # Lưu kết quả vào file CSV mới
    filtered_df.to_csv(output_file_path, index=False, encoding='utf-8-sig')

    print(f"File {file_name} đã được lọc outliers và lưu vào file CSV")
    print("_"*50)
    print("-"*50)
```
